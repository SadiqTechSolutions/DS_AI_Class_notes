```python
# Import libraries - These are the tools we need for data visualization
# ---------------------------------------------------------------

# NumPy: For numerical operations and creating test data
import numpy as np

# Pandas: For working with tabular data (like Excel spreadsheets)
import pandas as pd

# Matplotlib: The main plotting library in Python
# 'plt' is the standard short name we use to refer to matplotlib.pyplot
import matplotlib.pyplot as plt

# Seaborn: A higher-level visualization library built on matplotlib
# 'sns' is the standard short name for seaborn
import seaborn as sns


# Set the visual style for our plots
# ---------------------------------------------------------------
# This adds a light grid to our plots, making them easier to read
# Other options include: 'darkgrid', 'whitegrid', 'dark', 'white', and 'ticks'
sns.set_style("whitegrid")


# Create sample data to practice with
# ---------------------------------------------------------------

# Set a random seed to make sure we all get the same random numbers
# This is important for teaching and reproducibility
np.random.seed(42)

# Create 30 dates starting from January 1, 2023
# This will be our x-axis for time series plots
dates = pd.date_range('2023-01-01', periods=30)

# Create sales data with random values between 50 and 200
# We add a linear trend (0 to 40) to make the data more interesting
sales = np.random.randint(50, 200, size=30) + np.linspace(0, 40, 30)

# Create temperature data using a normal (bell curve) distribution
# Average temperature is 70 with a standard deviation of 10
temperatures = np.random.normal(70, 10, 30)

# Create category labels (A, B, C, D) randomly assigned
categories = np.random.choice(['A', 'B', 'C', 'D'], 30)

# Create region labels randomly assigned
regions = np.random.choice(['North', 'South', 'East', 'West'], 30)


# Organize all our data into a Pandas DataFrame
# ---------------------------------------------------------------
# A DataFrame is like a table or spreadsheet in Python
df = pd.DataFrame({
    'Date': dates,           # Column name: Date, values from our dates list
    'Sales': sales,          # Column name: Sales, values from our sales list
    'Temperature': temperatures,  # Column name: Temperature, values from our temperatures list
    'Category': categories,   # Column name: Category, values from our categories list
    'Region': regions         # Column name: Region, values from our regions list
})

# Show a preview of our data to verify everything looks correct
print("Sample DataFrame:")
df.head()  # The .head() method shows the first 5 rows of the DataFrame
```

```python
# First, create a figure with a specific size (width: 10 inches, height: 6 inches)
# This helps us control how big our plot will be
plt.figure(figsize=(10, 6))

# Create the actual line plot
# We use df['Date'] for x-axis values and df['Sales'] for y-axis values
# This draws a line connecting all our data points
plt.plot(df['Date'], df['Sales'])

# Add a title to explain what the plot shows
plt.title('Daily Sales per Day')

# Label the x-axis (horizontal axis)
plt.xlabel('Dates')

# Label the y-axis (vertical axis)
plt.ylabel('Sale')

# Add grid lines to make it easier to read values
plt.grid(True)

# Adjust the layout to make sure everything fits nicely
plt.tight_layout()

# Display the plot
# Without this command, the plot won't appear!
plt.show()
```

```python
# Create another figure with the same size
plt.figure(figsize=(10, 6))

# Plot the first half of our data (first 15 days)
# 'b-' means: blue color with solid line
# label='First Half' will be used in the legend
plt.plot(df['Date'][:15], df['Sales'][:15], 'b-', label='First Half')

# Plot the second half of our data (last 15 days)
# 'r--' means: red color with dashed line
# label='Second Half' will be used in the legend
plt.plot(df['Date'][15:], df['Sales'][15:], 'r--', label='Second Half')

# Add a title
plt.title('Sales Comparison')

# Label the axes
plt.xlabel('Date')
plt.ylabel('Sales')

# Add a legend that explains what each line represents
# This uses the labels we defined in the plt.plot() functions
plt.legend()

# Add grid lines
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
# Create a figure with specific size
plt.figure(figsize=(10, 6))

# Create the bar chart
# We use category_sales.index (the category names: A, B, C, D) for the x-axis
# We use category_sales.values (the sales totals) for the y-axis (bar heights)
plt.bar(category_sales.index, category_sales.values)

# Add a title
plt.title('Total Sales by Category')

# Label the axes
plt.xlabel('Category')
plt.ylabel('Total Sales')

# Add horizontal grid lines only (axis='y')
# This makes it easier to read the sales values
plt.grid(True, axis='y')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Here we group by both 'Category' AND 'Region'
# The 'unstack()' function reshapes the data so we can plot multiple bars per category
category_region = df.groupby(['Category', 'Region'])['Sales'].sum().unstack()

# Create a figure
plt.figure(figsize=(12, 6))

# Create a grouped bar chart
# The 'kind='bar'' parameter tells pandas to make a bar plot
# Each category will have multiple bars, one for each region
category_region.plot(kind='bar')

# Add a title
plt.title('Sales by Category and Region')

# Label the axes
plt.xlabel('Category')
plt.ylabel('Sales')

# Add a legend with a title to explain the colors
plt.legend(title='Region')

# Add horizontal grid lines
plt.grid(True, axis='y')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Create a figure with specific size
plt.figure(figsize=(10, 6))

# Create a scatter plot
# Each point represents one day's data
# x-axis: Temperature values
# y-axis: Sales values
# This helps us see if there's a relationship between temperature and sales
plt.scatter(df['Temperature'], df['Sales'])

# Add a title that describes the relationship we're examining
plt.title('Sales vs. Temperature')

# Label the axes
plt.xlabel('Temperature')
plt.ylabel('Sales')

# Add grid lines to make it easier to read values
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python

# Create a figure
plt.figure(figsize=(10, 6))

# Get the unique categories from our data (A, B, C, D)
# This creates a list of the different categories
categories = df['Category'].unique()

# Define a color for each category
colors = ['blue', 'green', 'red', 'purple']

# Loop through each category and plot its points with a unique color
# The enumerate function gives us both the index (i) and the value (category)
for i, category in enumerate(categories):
    # Create a mask (filter) that's True only for rows matching this category
    # For example, if category='A', this is True where df['Category'] is 'A'
    mask = df['Category'] == category
    
    # Plot just the points for this category with its specific color
    # df.loc[mask, 'Temperature'] gets temperatures only for this category
    # df.loc[mask, 'Sales'] gets sales only for this category
    plt.scatter(df.loc[mask, 'Temperature'],   # x-values (Temperature) 
                df.loc[mask, 'Sales'],         # y-values (Sales)
                c=colors[i],                   # Color from our colors list
                label=category,                # Label for the legend
                alpha=0.7)                     # Transparency (0.7 = 70% opaque)

# Add a descriptive title
plt.title('Sales vs. Temperature by Category')

# Label the axes
plt.xlabel('Temperature')
plt.ylabel('Sales')

# Add a legend to explain what each color represents
# This uses the labels we defined in the plt.scatter() function
plt.legend()

# Add grid lines
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# EXAMPLE 1: Simple histogram to show data distribution
# ---------------------------------------------------------------

# Create a figure with specific size
plt.figure(figsize=(10, 6))

# Create a histogram with our sales data
# A histogram divides the data into ranges (bins) and counts how many values fall into each bin
# bins=10 means we'll divide our data into 10 equal ranges
# edgecolor='black' adds a black outline to each bar to make them more distinct
plt.hist(df['Sales'], bins=10, edgecolor='black')

# Add a title explaining what we're looking at
plt.title('Distribution of Sales')

# Label the x-axis (the sales values)
plt.xlabel('Sales')

# Label the y-axis (how many times each value range occurs)
plt.ylabel('Frequency')

# Add horizontal grid lines to make it easier to read frequency values
plt.grid(True, axis='y')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
