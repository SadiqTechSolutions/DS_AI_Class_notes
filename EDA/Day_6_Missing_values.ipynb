```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'age': [25, np.nan, 35, 28, np.nan, 32],
    'salary': [50000, 60000, np.nan, 55000, 65000, 70000],
    'department': ['Engineering', 'Marketing', 'Engineering', np.nan, 'HR', 'Marketing'],
    'experience': [2, np.nan, 7, 3, 5, np.nan]
}
df = pd.DataFrame(data)

# Check for missing values (returns a boolean DataFrame)
print("Missing Value Check (True means missing):")
print(df.isna())

# Count missing values per column
print("\nMissing Value Count per Column:")
print(df.isna().sum())

# Calculate the percentage of missing values
print("\nPercentage of Missing Values:")
print((df.isna().sum() / len(df)) * 100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# A heatmap provides a clear visual of wherce missing values are located
plt.figure(figsize=(10, 6))
sns.heatmap(df.isna(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Heatmap of Missing Data')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

```python
# Sample data: A series of financial transactions
data = {
    'transaction_id': range(1, 21),
    'amount': [100, 105, 98, 110, 102, 107, 5000, 95, 103, 108, 
               101, 99, 112, 106, -200, 104, 109, 100, 111, 105]
}
df_trans = pd.DataFrame(data)
```

```python
df_trans.head()
```

```python
# A box plot is excellent for showing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_trans['amount'], y = df_trans['transaction_id'])
plt.title('Box Plot of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Transaction Id')
plt.show()
```

```python
# A scatter plot can also reveal outliers in the context of other variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='transaction_id', y='amount', data=df_trans)
plt.title('Scatter Plot of Transaction Amounts')
plt.xlabel('Transaction ID')
plt.ylabel('Amount')
plt.show()
```
