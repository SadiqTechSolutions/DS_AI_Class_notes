```python
# 1. Creating series in pandas with default index
list_1 = [1,2,3.0,4,4,44,444,44444,66666]
series_1 = pd.Series(list_1)
print(series_1)
series_1.dtype
```

```python
# Creating series in pandas with custom index
series_2 = pd.Series([1,3,4444444,8888], index=['aaa','bbbb','cccc','dddd'])
print(series_2)
series_2.dtype
series_2.shape
```

```python
# 3. Creating Dataframes in pandas

dict_1 = {
    'abc' : [1,2,3,4],
    'def' : [5,6,7,8],
    'ghi' : [9,10,11,12]
}

df_1 = pd.DataFrame(dict_1, index = ['place1', 'place2', 'place3', 'place4'])
print(df_1)
print(df_1.shape)
```
